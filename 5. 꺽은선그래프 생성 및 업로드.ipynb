```python
import datetime
print(datetime.datetime.now())

now = datetime.datetime.now()
now = "{}-{}".format(now.month, now.day)

day = datetime.date(2018, 11, 20)
all_keyword = list()

from collections import Counter

while True :
    target = "{}-{}".format(day.month, day.day)
    if target == now : break
    f = open("./per_day_keyword/(keyword)" + target + ".txt", mode = "r", encoding="utf-8")
    text = f.read()
    exec("text = " + text)
    f.close()
    
    count = Counter(text)
    keywords = count.most_common(50)
    keys = [ item[0] for item in keywords ]
    
    all_keyword += keys
    all_keyword = list(set(all_keyword))
    day += datetime.timedelta(1)

# 버킷 함수 지정
from google.cloud import storage
def upload_blob(bucket_name, source_file_name, destination_blob_name):
    """Uploads a file to the bucket."""
    storage_client = storage.Client()
    bucket = storage_client.get_bucket(bucket_name)
    blob = bucket.blob(destination_blob_name)
    blob.upload_from_filename(source_file_name)
    print('File {} uploaded to {}.'.format(
        source_file_name,
        destination_blob_name))

# 꺽은선 그래프
import matplotlib.pyplot as plt
import matplotlib.legend as legend

# 키워드에 대해서 반복함.
for keyword in all_keyword :
    day = datetime.date(2018, 11, 20)
    counts = list() # 세로축 : 날짜 별 언급된 횟수
    terms = list() # 가로축 : 기간
    while True :
        target = "{}-{}".format(day.month, day.day)
        if target == now : break
        f = open("./per_day_keyword/(keyword)" + target + ".txt", mode = "r", encoding="utf-8")
        text = f.read()
        f.close()
        exec("text = " + text)
        terms.append(target)
        counts.append(text.count(keyword))
        day += datetime.timedelta(1)
        
    gab,=plt.plot(terms, counts, 'bo-')
    
    plt.legend([gab], ['RAB'], loc=1)
    plt.xlabel(keyword)
    plt.ylabel('Number of References')
    plt.grid(True)
    plt.savefig("./line-graph/" + keyword + ".jpg")
    plt.show()
    plt.gcf().clear()
    
    bucket_name = 'gunmin_test'
    source_file_name = "./line-graph/" + keyword + ".jpg"
    destination_blob_name = "line-graph/" + keyword + ".jpg"
    
    upload_blob(bucket_name, source_file_name, destination_blob_name)
    
print(datetime.datetime.now())
```
